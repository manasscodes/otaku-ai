```python
def score_to_sentiment(score):
    if score >= 8:
        return "Positive"
    elif score >= 5:
        return "Neutral"
    else:
        return "Negative"

df["sentiment"] = df["score"].apply(score_to_sentiment)
```

# Dataset Audit Findings

## Finding 1
68.25% of reviews are exactly 2000 characters long, suggesting truncation or a platform review limit.

## Finding 2
Review length appears to have little relationship with sentiment.
Average review lengths are similar across Positive, Neutral, and Negative classes.

## Finding 3
The dataset is mildly imbalanced:
- Positive: 849
- Neutral: 390
- Negative: 251

## Finding 4
episodes_watched is completely missing and is unlikely to be useful.
